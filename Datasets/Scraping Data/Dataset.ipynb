```python
# Define the files and their corresponding 'Type' labels
data_info = [
    ('ielts_academic_task1.csv', 'academic'),
    ('ielts_general_task1.csv', 'general'),
    ('ielts_task2.csv', 'task 2')
]

dfs = []

for filename, label in data_info:
    df = pd.read_csv(filename)
    df['Type'] = label
    dfs.append(df)

combined_df = pd.concat(dfs, ignore_index=True)

# Reorder columns to put 'Type' at the beginning
cols = ['Type'] + [col for col in combined_df.columns if col != 'Type']
combined_df = combined_df[cols]

# Save the combined dataset
combined_df.to_csv('combined_ielts_essays.csv', index=False, encoding='utf-8-sig')

print("Files combined successfully into 'combined_ielts_essays.csv'")
```

```python
import pandas as pd
import re

def normalize_text(text):
    """
    Removes newlines (bullet points), tabs, and extra spaces,
    converting the text into a single continuous paragraph.
    """
    if not isinstance(text, str):
        return text
    text = re.sub(r'\s+', ' ', text)
    return text.strip()

def clean_essay_with_signature(text):
    """
    Removes the username/date signature from the end and then normalizes whitespace.
    """
    if not isinstance(text, str):
        return text
    
    lines = text.strip().split('\n')
    date_pattern = re.compile(r'[A-Z][a-z]{2}\s\d{1,2},\s\d{4}')
    
    if len(lines) >= 2 and date_pattern.search(lines[-1]):
        cleaned_lines = lines[:-2]
    else:
        cleaned_lines = lines
    
    return normalize_text(" ".join(cleaned_lines))

df = pd.read_csv('combined_ielts_essays.csv')

df['Question'] = df['Question'].apply(normalize_text)

df['Essay_Text'] = df['Essay_Text'].apply(clean_essay_with_signature)

df['Type'] = df['Type'].str.lower().str.strip().replace('task2', 'task 2')

df.to_csv('combined_ielts_essays_fixed.csv', index=False, encoding='utf-8-sig')

print("Process Complete! 'Question' and 'Essay_Text' are now standardized.")
```

```python
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
df = pd.read_csv('combined_ielts_essays.csv')

score_cols = ['Overall_Score', 'Task_Achievement', 'Coherence_Cohesion', 'Lexical_Resource', 'Grammar_Range']

# 1. Distribution of Type
plt.figure(figsize=(8, 5))
sns.countplot(data=df, x='Type', palette='viridis')
plt.title('Distribution of Essay Types')
plt.xlabel('Type')
plt.ylabel('Count')
plt.savefig('distribution_type.png')
plt.close()

# 2. Distributions of Scores
fig, axes = plt.subplots(len(score_cols), 1, figsize=(10, 20))
for i, col in enumerate(score_cols):
    sns.countplot(data=df, x=col, ax=axes[i], palette='magma')
    axes[i].set_title(f'Distribution of {col}')
    axes[i].set_xlabel('Score')
    axes[i].set_ylabel('Count')

plt.tight_layout()
plt.savefig('distribution_scores.png')
plt.close()

# 3. Distribution of Essay Word Count
df['Word_Count'] = df['Essay_Text'].str.split().str.len()
plt.figure(figsize=(10, 6))
sns.histplot(df['Word_Count'], bins=30, kde=True, color='blue')
plt.title('Distribution of Essay Word Counts')
plt.xlabel('Word Count')
plt.ylabel('Frequency')
plt.savefig('distribution_word_count.png')
plt.close()

print("Plots generated: distribution_type.png, distribution_scores.png, distribution_word_count.png")
print(df[score_cols + ['Word_Count']].describe())
```
